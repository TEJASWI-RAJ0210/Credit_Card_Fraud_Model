### Applying Standard Scaler On Amount and Time

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler=StandardScaler()

df["Amount_scaled"]=scaler.fit_transform(df[["Amount"]])
df["Time_scaled"]=scaler.fit_transform(df[["Time"]])
```

```python
df.head()
```

```python
df.drop(["Amount","Time"],axis=1,inplace=True)
```

```python
df.head()
```

```python
X=df.drop(["Class"],axis=1)
y=df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42,stratify=y)
```

```python
print("Train fraud rate:", y_train.mean().round(4))
print("Test fraud rate: ", y_test.mean().round(4))
print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
```

### Imbalance Handling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
model_lr=LogisticRegression(class_weight='balanced', max_iter=1000)
model_rf=RandomForestClassifier(class_weight='balanced', n_estimators=100)
```

```python
X_train
```

```python
model_lr.fit(X_train,y_train)
```

```python
model_rf.fit(X_train,y_train)
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42,k_neighbors=5)
X_smote,y_smote=smote.fit_resample(X_train,y_train)
print("Before SMOTE:", y_train.value_counts().to_dict())
print("After SMOTE: ", pd.Series(y_smote).value_counts().to_dict())
```

```python
smote=SMOTE(random_state=42,k_neighbors=5)
X_smote_test,y_smote_test=smote.fit_resample(X_test,y_test)
print("Before SMOTE:", y_test.value_counts().to_dict())
print("After SMOTE: ", pd.Series(y_smote_test).value_counts().to_dict())
```

```python
import xgboost as xgb

neg = (y_train == 0).sum()
pos = (y_train == 1).sum()
spw = neg / pos        # ≈ 577

xgb_model = xgb.XGBClassifier(
    scale_pos_weight=spw,
    n_estimators=200,
    max_depth=4,
    learning_rate=0.1,
    use_label_encoder=False,
    eval_metric='aucpr',    # use PR-AUC not accuracy
    random_state=42
)
xgb_model.fit(X_train, y_train)
```

```python
y_pred_lr=model_lr.predict(X_test)
y_pred_rf=model_rf.predict(X_test)
y_pred_xg=xgb_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
confusion_matrix(y_test,y_pred_lr)
print(classification_report(y_test,y_pred_lr))
```

```python
confusion_matrix(y_test,y_pred_rf)
print(classification_report(y_test,y_pred_rf))
confusion_matrix(y_test,y_pred_xg)
print(classification_report(y_test,y_pred_xg))
```

```python
print("Random Forest",confusion_matrix(y_test,y_pred_rf))
print("XGBoost",confusion_matrix(y_test,y_pred_xg))
```

```python
from sklearn.metrics import average_precision_score,PrecisionRecallDisplay
y_proba_rf = model_rf.predict_proba(X_test)[:, 1]
y_proba_xg = xgb_model.predict_proba(X_test)[:, 1]
```

```python
print("PR_AUC_rf",average_precision_score(y_test,y_proba_rf))
print("PR_AUC_xg",average_precision_score(y_test,y_proba_xg))
```

```python
PrecisionRecallDisplay.from_predictions(y_test, y_proba_rf)
plt.title('Precision-Recall curve Random Forest')
plt.show()
```

```python
PrecisionRecallDisplay.from_predictions(y_test, y_proba_xg)
plt.title('Precision-Recall curve XGBoost')
plt.show()
```

### For credit card fraud detection, I would select XGBoost because:

- It catches more fraudulent transactions (81% vs 73% recall).
- It has the highest PR-AUC, which is one of the best metrics for highly imbalanced datasets.
- Missing frauds is usually more expensive than investigating a few extra alerts.

### SHAP EXPLAINABILITY

```python
import shap
```

```python
# Always use TreeExplainer for XGBoost — it's fast and exact
explainer = shap.TreeExplainer(xgb_model)

# Compute SHAP values on the test set
# Use a sample if test set is large (speeds things up)
X_test_sample = X_test.sample(n=1000, random_state=42)
shap_values = explainer(X_test_sample)

print(f"SHAP values shape: {shap_values.shape}")
```

```python
plt.figure()
shap.summary_plot(
    shap_values,
    X_test_sample,
    plot_type="dot",        # beeswarm — best for understanding feature impact
    max_display=15,         # show top 15 features
    show=False
)
plt.title("SHAP summary — feature impact on fraud prediction")
plt.tight_layout()
plt.savefig("shap_summary.png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
shap.summary_plot(
    shap_values,
    X_test_sample,
    plot_type="bar",        # mean absolute SHAP — easier to explain to non-ML people
    max_display=15
)
```

```python
# Find actual fraud cases in your test sample
fraud_indices = X_test_sample[y_test[X_test_sample.index] == 1].index
legit_indices = X_test_sample[y_test[X_test_sample.index] == 0].index

# Pick the first fraud transaction
fraud_pos = X_test_sample.index.get_loc(fraud_indices[0])

plt.figure()
shap.plots.waterfall(
    shap_values[fraud_pos],
    max_display=15,
    show=False
)
plt.title("Why did the model flag THIS transaction as fraud?")
plt.tight_layout()
plt.savefig("shap_waterfall_fraud.png", dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Pick the first legitimate transaction
legit_pos = X_test_sample.index.get_loc(legit_indices[0])

plt.figure()
shap.plots.waterfall(
    shap_values[legit_pos],
    max_display=15,
    show=False
)
plt.title("Why did the model clear THIS transaction as legitimate?")
plt.tight_layout()
plt.savefig("shap_waterfall_legit.png", dpi=150, bbox_inches='tight')
plt.show()
```
