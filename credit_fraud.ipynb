```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('creditcard.csv')
df.head()
```

```python
sns.barplot(x=df["Class"].value_counts().index,y=df["Class"].value_counts().values)
plt.xlabel("Class")
```

```python
sns.boxplot(x=df["Class"],y=df["Amount"])
```

```python
corr_matrix = df[["V1", "V2", "V3", "V4","V5","V6","V7","V8","V9","V10" ,"Class"]].corr()
plt.figure(figsize=(14,10))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
corr_matrix2 = df[["V11", "V12", "V13", "V14","V15","V16","V17","V18","V19","V20" ,"Class"]].corr()
plt.figure(figsize=(14,10))
sns.heatmap(corr_matrix2, annot=True, cmap="coolwarm")
```

```python
corr_matrix3 = df[["V21", "V22", "V23", "V24","V25","V26","V27","V28","Amount","Time" ,"Class"]].corr()
plt.figure(figsize=(14,10))
sns.heatmap(corr_matrix3, annot=True, cmap="coolwarm")
```

### Some Findings
- V4,V11,V12,V14,V17-Strong relation
- Fraud occurred for less than 5000 amount

```python
sns.violinplot(data=df,x="Class",y="Time")
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12,6), sharex=True)

df[df['Class']==0]['Time'].plot(kind='density', ax=axes[0],
    title='Legitimate — transaction density over time', color='#1D9E75')
df[df['Class']==1]['Time'].plot(kind='density', ax=axes[1],
    title='Fraud — transaction density over time', color='#D85A30')

plt.xlabel('Time (seconds)')
plt.tight_layout()
plt.show()
```

```python
top_features = ['V4', 'V11', 'V12', 'V14', 'V17']
fig, axes = plt.subplots(1, 5, figsize=(18,4))

for i, feat in enumerate(top_features):
    # df[df['Class']==0][feat].plot(kind='hist', ax=axes[i],
    #     alpha=0.6, color='#1D9E75', label='Legit', bins=50)
    df[df['Class']==1][feat].plot(kind='hist', ax=axes[i],
        alpha=0.6, color='#D85A30', label='Fraud', bins=50)
    axes[i].set_title(feat)
    axes[i].legend(fontsize=8)

plt.tight_layout()
plt.show()
```

```python
corr_with_fraud = df.corr()['Class'].drop('Class').sort_values()

print("Most NEGATIVE correlation (strong fraud signal):")
print(corr_with_fraud.head(10))

print("\nMost POSITIVE correlation (strong fraud signal):")
print(corr_with_fraud.tail(10))
```
